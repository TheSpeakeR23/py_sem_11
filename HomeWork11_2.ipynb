```python
from sympy import *
from sympy.plotting import plot

x = Symbol("x")
f = 2 * x ** 3 + 2 * x ** 2 - 18 * x - 18
f
```

```python
# 1. Определить корни
#    Нули функции
solveset(f, x) 
```

```python
# 2. Найти интервалы, на которых функция возрастает
# 3. Найти интервалы, на которых функция убывает

f_diff = [-oo, oo]
f_diff[1:1] = solve(diff(f), x)
f_diff
```

```python
incr_list = []
decr_list = []

for i in range(1, len(f_diff)):
    val = is_increasing(f, Interval.open(f_diff[i - 1], f_diff[i]))
    if not val:
        incr_list.append(f"[{f_diff[i - 1]}, {f_diff[i]}]")
    else:
        decr_list.append(f"[{f_diff[i - 1]}, {f_diff[i]}]")

print(f"Убывает на интервалах:", *decr_list, sep="\n")
print(f"Возрастает на интервалах:", *incr_list, sep="\n")
```

```python
# 4. Построить график

plot(f, (x, -4, 4), legend=True)
```

```python
# 5. Вычислить вершину
#    Экстремумы функции

ext_list = solve(diff(f), x)

for i in ext_list:
    res = f.subs(x, i)
    if res < 0:
        print(f"Точка минимума: (x:{i}, y:{res})")
    elif res > 0:
        print(f"Точка максимума: (x:{i}, y:{res})")
```

```python
# 6. Определить промежутки, на котором f > 0
# 7. Определить промежутки, на котором f < 0
#    Знакопостоянства функции


lst = [-oo, oo]
lst[1:1] = solve(f, x)

inc = []
red = []
for i in range(1, len(lst)):
    temp_1, temp_2 = lst[i-1], lst[i]
    temp = is_increasing(f, Interval.open(temp_1, temp_2))
    if temp:
        inc.append(f'{temp_1},{lst[i]}')
    else:
        red.append(f'{temp_1},{lst[i]}')
print('Возрастает: ', *inc, sep = '\n')
print('Убывает:', *red, sep = '\n')
```
